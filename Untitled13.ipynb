```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import roc_auc_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import os
```

```python
os.chdir("C:\\Users\\DELL\\Downloads")
df = pd.read_csv("fraud_data.csv")
```

```python
df.isnull()
```

```python
target_col = None
for t in ['is_fraud', 'fraud', 'label', 'target', 'Class', 'y']:
    if t in df.columns:
        target_col = t
        break
if target_col is None:
    target_col = df.columns[-1] 
```

```python
df.info()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['Amount'].fillna(df['Amount'].mean(), inplace=True)
df['CardHolderAge'].fillna(df['CardHolderAge'].mean(), inplace=True)
df['Location'].fillna(df['Location'].mode()[0], inplace=True)
```

```python
X = df.drop(columns=[target_col])
y = df[target_col]
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
imputer = SimpleImputer(strategy="mean")
X = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)
```

```python
df.isnull().sum()
```

```python
df.head(20)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
log_reg = LogisticRegression(max_iter=1000, class_weight='balanced')
log_reg.fit(X_train, y_train)
y_pred_prob_lr = log_reg.predict_proba(X_test)[:, 1]
roc_auc_lr = roc_auc_score(y_test, y_pred_prob_lr)
print(f"Logistic Regression ROC-AUC: {roc_auc_lr:.4f}")
print(classification_report(y_test, log_reg.predict(X_test)))
```

```python
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf.fit(X_train, y_train)
y_pred_prob_rf = rf.predict_proba(X_test)[:, 1]
roc_auc_rf = roc_auc_score(y_test, y_pred_prob_rf)
print(f"\nRandom Forest ROC-AUC: {roc_auc_rf:.4f}")
print(classification_report(y_test, rf.predict(X_test)))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

# --- ROC Curve Plot ---
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_pred_prob_lr)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_prob_rf)

plt.figure(figsize=(8, 6))
plt.plot(fpr_lr, tpr_lr, label=f"Logistic Regression (AUC = {roc_auc_lr:.4f})")
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC = {roc_auc_rf:.4f})")
plt.plot([0, 1], [0, 1], 'k--', label="Random Guess")  # diagonal line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Fraud Detection")
plt.legend()
plt.grid(True)
plt.show()
```

```python
better_model = "Random Forest" if roc_auc_rf > roc_auc_lr else "Logistic Regression"
print(f"\nâœ… Recommended Model: {better_model} (based on higher ROC-AUC)")
```
